## $\textbf{1. The equations of motion}$

Writing down the equations of motion in latex for this problem will probably be a "Goat-Hell", but I am confident we will make it through. The position of the counterweight is given by 
$$
\vec{r_1} = 
\begin{bmatrix}
l_1sin(\theta) \\
-l_1cos(\theta) + l_3
\end{bmatrix}
$$

The velocity is then obtained by taking the time derivative:
$$
\vec{v_1} = \frac{d\vec{r_1}}{dt} = 
\begin{bmatrix}
l_1cos(\theta)\dot{\theta} \\
l_1sin(\theta)\dot{\theta}
\end{bmatrix}
$$

The position of the projectile is given by
$$
\vec{r_{2b}} = 
\begin{bmatrix}
-l_2sin(\theta) - l_4sin(\psi - \theta) \\
l_2cos(\theta) + l_3 - l_4cos(\psi - \theta)
\end{bmatrix}
$$

In the same manner as for the counterweight we find the velocity of the projectile: 

$$
\vec{v_{2b}} = \frac{d\vec{r_{2b}}}{dt} =  
\begin{bmatrix}
-l_2cos(\theta)\dot{\theta} - l_4cos(\psi - \theta) (\dot{\psi} - \dot{\theta}) \\
-l_2sin(\theta)\dot{\theta} + l_4sin(\psi + \theta) (\dot{\psi} - \dot{\theta})
\end{bmatrix}
$$

We will now use these expressions to construct the lagrangian for the trebuchet, where $L = T - V$. The kinetic energy is given by 

$$
T = \sum_{i=1}^{2}\frac12 m_iv_i^2 = \sum_{i=1}^{2}\frac12 m_i(\dot{x_i}^2 + \dot{y_i}^2)
$$

Oh boy. What is there to do but to roll up my sleves and force myself through these painful calculations. Wish me luck...

$$
T = \frac12 m_1((l_1cos(\theta)\dot{\theta})^2 + (l_1sin(\theta)\dot{\theta})^2)
+ \frac12 m_2((-l_2cos(\theta)\dot{\theta} - l_4cos(\psi - \theta) (\dot{\psi} - \dot{\theta}))^2 + (-l_2sin(\theta)\dot{\theta} + l_4sin(\psi - \theta) (\dot{\psi} - \dot{\theta}))^2)
$$


As one sees, it is the velocity of the projectile which causes a severe headache. However, with the help of some trigonometric identities, the expression can be simplified. We here use the fact that 

$$
cos(\theta)cos(\psi - \theta) - sin(\theta)sin(\psi - \theta) = cos(\theta + \psi - \theta) = cos(\psi)
$$

and get 

$$
T = \frac12 (m_1l_1^2 + m_2l_2^2)\dot{\theta}^2 + \frac12 m_2l_4^2(\dot{\psi} - \dot{\theta})^2 + m_2l_2l_4\dot{\theta}(\dot{\psi} - \dot{\theta})cos(\psi)
$$

We can finally move out attention to the potential energy of the system, which is given by 

$$
V = \sum_{i=1}^{2}mgy_i
$$

where $y_1$ and $y_2$ are the heights of the counterweight and projectile respectively. 

This gives 

$$
V = g(-m_1l_1cos(\theta) + m_1l_3 + m_2l_2cos(\theta) + m_2l_3 - m_2l_4cos(\psi - \theta))
$$

And here we go! We now have all the ingredients to write down our lagrangian, which now is: 

$$
L = T - V = \frac12 (m_1l_1^2 + m_2l_2^2)\dot{\theta}^2 + \frac12 m_2l_4^2(\dot{\psi} - \dot{\theta})^2 + m_2l_2l_4\dot{\theta}(\dot{\psi} - \dot{\theta})cos(\psi) - gl_3(m_1 + m_2) + g(m_1l_1 - m_2l_2)cos(\theta) + gm_2l_4cos(\psi - \theta)
$$

The joy and relief of finding the lagrangian is short-lived, as we now move on to applying the Euler_Lagrange equations for our two generalized coordinates $\theta$ and $\psi$. 

To avoid errors, this will be done step by step: 

$$
\begin{align}
\frac{\partial{L}}{\partial{\theta}} &= -g(m_1l_1 - m_2l_2)sin(\theta) + gm_2l_4sin(\psi - \theta)\\
\frac{\partial{L}}{\partial{\dot{\theta}}} &= (m_1l_1^2 + m_2l_2^2)\dot{\theta} - m_2l_4^2(\dot{\psi} - \dot{\theta}) + m_2l_2l_4(\dot{\psi} - \dot{\theta})cos(\psi) - m_2l_2l_4\dot{\theta}cos(\psi) \\
\frac{d}{dt} \frac{\partial{L}}{\partial{\dot{\theta}}} &= (m_1l_1^2 + m_2l_2^2)\ddot{\theta} - m_2l_4^2(\ddot{\psi} - \ddot{\theta}) + m_2l_2l_4(\ddot{\psi} - \ddot{\theta})cos(\psi) - m_2l_2l_4(\dot{\psi} - \dot{\theta})sin(\psi) \dot{\psi} - m_2l_2l_4\ddot{\theta}cos(\psi) + m_2l_2l_4\dot{\theta}sin(\psi)\dot{\psi}\\
\frac{\partial{L}}{\partial{\psi}} &= -m_2l_2l_4\dot{\theta}(\dot{\psi} - \dot{\theta})sin(\psi) - gm_2l_4sin(\psi - \theta) \\
\frac{\partial{L}}{\partial{\dot{\psi}}} &= m_2l_4^2(\dot{\psi} - \dot{\theta}) + m_2l_2l_4\dot{\theta}cos(\psi) \\
\frac{d}{dt}\frac{\partial{L}}{\partial{\dot{\psi}}} &= m_2l_4^2(\ddot{\psi} - \ddot{\theta}) + m_2l_2l_4\ddot{\theta}cos(\psi) - m_2l_2l_4\dot{\theta}sin(\psi)\dot{\psi}
\end{align}
$$


We are now FINALLY ready to write down the equations of motion: 

$$ \begin{align}
&I:-g(m_1l_1 - m_2l_2)sin(\theta) + gm_2l_4sin(\psi - \theta)  - (m_1l_1^2 + m_2l_2^2)\ddot{\theta} + m_2l_4^2(\ddot{\psi} - \ddot{\theta}) - m_2l_2l_4(\ddot{\psi} - \ddot{\theta})cos(\psi) + m_2l_2l_4(\dot{\psi} - \dot{\theta})sin(\psi) \dot{\psi} + m_2l_2l_4\ddot{\theta}cos(\psi) - m_2l_2l_4\dot{\theta}sin(\psi)\dot{\psi} = 0 \\\\
&II: - m_2l_2l_4\dot{\theta}(\dot{\psi} - \dot{\theta})sin(\psi) - gm_2l_4sin(\psi - \theta) - m_2l_4^2(\ddot{\psi} - \ddot{\theta}) - m_2l_2l_4\ddot{\theta}cos(\psi) + m_2l_2l_4\dot{\theta}sin(\psi)\dot{\psi} = 0
\end{align}
$$

This can also be written as the following matrix equation: 

$$
\begin{bmatrix}
a_{11} & a_{12} \\
a_{12} & a_{22}
\end{bmatrix}
\begin{bmatrix}
\ddot{\theta} \\
\ddot{\psi}
\end{bmatrix}
=
\begin{bmatrix}
b_1 \\
b_2
\end{bmatrix}
$$

with the following matrix elements: 

$$\begin{align}
a_{11} &= -l_1^2m_1 - l_2^2m_2 + 2l_2l_4m_2cos(\psi) - l_4^2m_2
\\
a_{12} &= -l_4m_2(l_2cos(\psi) - l_4)
\\
a_{22} &= -l_4^2m_2
\\
b_1 &= gm_1l_1sin(\theta) - gm_2[l_2sin(\theta) + l_4sin(\psi-\theta)] - l_2l_4m_2(\dot{\psi} - 2\dot{\theta})sin(\psi)\dot{\psi}
\\
b_2 &= l_4m_2[gsin(\psi-\theta) - l_2sin(\psi)\dot{\theta}^2]
\end{align}
$$

That certanly was a handful, but fortunately we can now move on from this "goat-hell" latex. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate #Will use built-in RK45
import numpy.linalg
```

```python
#Some constants and initial conditions. 

m_1 = 2000
m_2 = 15
l_1 = 1.2
l_2 = 5.7
l_3 = 3.2
l_4 = 5
g = 9.81
theta0 = 0.7 * np.pi
thetad0 = 0
psi0 = theta0 - 0.5 * np.pi
psid0 = 0
```

## $\textbf{2. The RK4 method and ODE solver}$

RK4 method is defined by: 
$$
\begin{align}
k_1 &= \vec{f}(t_i, \vec\omega_i) \\
k_2 &= \vec{f}(t_i + h/2, \vec\omega_i + h\frac{k_1}{2}) \\
k_3 &= \vec{f}(t_i + h/2, \vec\omega_i + h\frac{k_2}{2}) \\
k_4 &= \vec{f}(t_i + h, \vec\omega_i hk_3) \\
\vec\omega_{i+1} &= \vec\omega_i + \frac{h}{6}(k_1 + 2k_2 + 2k_3 + k_4)
\end{align}
$$


```python

#The rk4 method
def rk4_method(f, t, y, h):
    k1 = f(t, y)
    k2 = f(t + h/2, y + h*k1/2)
    k3 = f(t + h/2, y + h*k2/2)
    k4 = f(t + h, y + h*k3)

    return y +  h/6*(k1 + 2*k2 + 2*k3 + k4)
```

```python

def projectile_trajectory(theta_arr, psi_arr): 
    x_pos = -l_2 * np.sin(theta_arr) - l_4 * np.sin(psi_arr - theta_arr)
    y_pos = l_2 * np.cos(theta_arr) + l_3 - l_4 * np.cos(psi_arr - theta_arr)
    return x_pos, y_pos

def counterWeight_trajectory(theta_arr, psi_arr): 
    x_pos = l_1 * np.sin(theta_arr)
    y_pos = - l_1 * np.cos(theta_arr) + l_3
    return x_pos, y_pos

def projectile_velocity(theta_arr, psi_arr, thetad_arr, psid_arr): 
    x_vel = -l_2 * np.cos(theta_arr) * thetad_arr - l_4 * np.cos(psi_arr - theta_arr) * (psid_arr - thetad_arr)
    y_vel = -l_2 * np.sin(theta_arr) * thetad_arr + l_4 * np.sin(psi_arr - theta_arr) * (psid_arr - thetad_arr)
    return x_vel, y_vel
```

```python

def f(t, omega_vec): #Our function f
    theta = omega_vec[0]
    psi = omega_vec[1]
    thetad = omega_vec[2]
    psid = omega_vec[3]
    
    #
    A = np.array([[-l_1**2 * m_1 - l_2**2 * m_2 + 2 * l_2 * l_4 * m_2 * np.cos(psi) - l_4**2 * m_2, 
               -l_4 * m_2 * (l_2 * np.cos(psi) -l_4)], 
             [-l_4 * m_2 * (l_2 * np.cos(psi) -l_4), 
             -l_4**2 * m_2]])
    b = np.array([[g * m_1 * l_1 * np.sin(theta) - 
               g * m_2 * (l_2 * np.sin(theta) + l_4 * np.sin(psi - theta)) 
               - l_2 * l_4 * m_2 * (psid - 2 * thetad) * np.sin(psi) * psid],
              [l_4 * m_2 * (g * np.sin(psi - theta) - l_2 * np.sin(psi) * thetad**2)]])
   
    new_arr = np.transpose(np.linalg.inv(A) @ b) #this matrix product to obtain the values of psi and theta (doubledot)
    #Return f-array consisting the single and double derivatives of theta and psi
    return np.array([thetad, psid, new_arr[0][0], new_arr[0][1]])

def f_oblique(t, omega_vec): 
    x = omega_vec[0]
    y = omega_vec[1]
    dx = omega_vec[2]
    dy = omega_vec[3]
    
    return np.array([dx, dy, 0, -g])
```

```python
def my_personal_ode_solver(f,t0,tend, omega_vec_0, h, method, release_angle = 0, release = True): 
    
    #Initalizing arrays 
    omega_vals = np.array([omega_vec_0])
    t_vals = np.array([t0])
    
    omega_vec = omega_vec_0
    
    new_vals = omega_vec
    
    t = 0
    
    while tend - t > 1.e-10:
        #We always will include the last point tend.
        if t + h > tend:
            h = tend- t

        if release == False or new_vals[0] > release_angle: #Projectile not released
        #calculates the new step given a method
            t += h
        
            new_vals = method(f, t, new_vals, h)
            omega_vals = np.concatenate((omega_vals, np.array([new_vals])))
            t_vals = np.append(t_vals, t)
        
        elif release and new_vals[0] <= release_angle: #projectile released
            #I wish to keep the information from 
            t_omega = omega_vals
            t_time = t_vals
            
            #Gather information from above
            x_pos, y_pos = projectile_trajectory(omega_vals[:,0], omega_vals[:,1])
            x_vel, y_vel = projectile_velocity(omega_vals[:,0], omega_vals[:,1], omega_vals[:,2], omega_vals[:,3])
            omega_vals = np.array([x_pos[-1], y_pos[-1], x_vel[-1], y_vel[-1]])
            new_vals  = omega_vals
            
            y = y_pos[-1]
            
            omega_vals = np.array([omega_vals]) #Dont even ask...
            
            while y > 0:
                t += h
                
                new_vals = method(f_oblique, t, new_vals, h)
                
                omega_vals = np.concatenate((omega_vals, np.array([new_vals])))
                
                np.append(t_vals, t)
                
                y = new_vals[1]
                
            return t_vals, omega_vals, x_pos, y_pos
       
    return  t_vals, omega_vals
```

## $\textbf{5. Task c, the critical hit!}$

After investigating the influence of the release angle and the initial conditions for $\theta$ and $\psi$, we will now find the best release angle in terms of distance traveled by the projectile and impact energy. Here we are back with the unstoppable duo $\theta_0 = 0.7\pi$ and $\psi_0 = \theta_0 - 0.5\pi$. Our strategy will here be to iterate through 100 angles in the interval $\theta \in [0^{\circ}, 29.83^{\circ}]$. Which angle is the best? And what is the worst possible release angle?
```python
def impact_energy(x_vel, y_vel): 
    return 0.5 * m_2 * (x_vel**2 + y_vel**2)

def distance(x_arr): 
    return x_arr[-1]
```
